```python
class Forma: #Definindo a classe forma
      def __init__(self, side1, side2): #Iniciando os atributos
          self.side1 = side1 #Dando valores aos metódos
          self.side2 = side2 #Dando valores aos metódos
      def get_area(self): #Ganhando um metódo
          return self.side1 * self.side2 #Calculando area
      def __str__(self): #recebendo tipo str
          return f'The area of this {self.__class__.__name__} is: {self.get_area()}' #Imprimindo resposta
```

```python
class Retangulo (Forma) : #Definindo a classe retangulo
  pass
```

```python
class Quadrado (Retangulo):  #Definindo a classe quadrado
  def __init__(self, side):  #Iniciando os atributos
      super().__init__(side, side) #Compactando um valor dentro de uma super lista
```

```python
class Triangulo(Retangulo):  #Definindo a classe triangulo
      def __init__(self, base, height):  #Iniciando os atributos 
          super().__init__(base, height) #Compactando um valor dentro de uma super lista
      def get_area(self): #Ganhando um metódo
          area = super().get_area()
          return area / 2 #Imprimindo resposta
```

```python
class Circulo(Forma):  #Definindo a classe circulo
      def __init__(self, radius):  #Iniciando os atributos
          self.radius = radius
      def get_area(self):  #Ganhando um metódo
          return pi * (self.radius ** 2) #Imprimindo resposta
```

```python
class Hexagono(Quadrado):  #Definindo a classe hexagono
      def get_area(self):  #Ganhando um metódo
          return (3 * sqrt(3) * self.side1 ** 2) / 2 #Imprimindo resposta
```

```python
rec = Retangulo(1, 2) 
print(rec)
# A área do retângulo é: 2

sqr = Quadrado(4)
print(sqr)
# A área do quadradado é: 16

tri = Triangulo(2, 3)
print(tri)
# A área do triângulo é:  3.0

cir = Circulo(4)
print(cir)
# A área do circulo é:  50.26548245743669

hex = Hexagono(3)
print(hex)
# A área do hexagono é:  23.382685902179844

breakpoint()
```
