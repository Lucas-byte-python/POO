```python
class Cookie: #Definindo a classe cookie
      def __init__(self, name, shape, chips='Chocolate'):#Iniciando atributos
        self.name = name #Definindo atributo
        self.shape = shape #Definindo atributo
        self.chips = chips #Definindo atributo
      def bake(self): #definido o metódo 'bake' dentro da classe Cookie
        print(f'This {self.name}, is being baked with the shape {self.shape} and chips of {self.chips}') #Imprimindo valores
        print('Enjoy your cookie!') #Imprimindo valores
```

```python
cookie1 = Cookie('Biscoito', 'Bolacha') #Definindo as instâncias do cookie 01
```

```python
cookie2 = Cookie('Awesome cookie', 'Star') #Definindo as instâncias do cookie 02
```

```python
isinstance(cookie1, Cookie) # Definido como True or False
isinstance(cookie1, int) # Definido como True or False
isinstance('a string', Cookie) # Definido como True or False
```
