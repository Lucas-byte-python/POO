```python
#@title
Numero_Secreto = 10.89
while True:
  numero = input("Número: ")
  try:
    numero = float(numero)
  except:
    print("Sorry!")
    continue
  if numero != Numero_Secreto:
    if numero > Numero_Secreto:
      print(numero, "Menor: ")
    elif numero < Numero_Secreto:
      print(numero, "Maior: ")
  else: 
    print("Acertou:", Numero_Secreto)
    break;

  
```
