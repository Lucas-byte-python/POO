**Testando Classe **

```python
class Cafe:
  def __init__(self, tipo, valor):
    self.tipo = tipo
    self.valor = float(valor)
  def check_valor(self, valDisponivel):
    if not isinstance(valDisponivel, (int, float)):
      print("Digite um valor int ou float")
      exit()
    if valDisponivel <= 0:
      print("Tu é pobre")
      exit(0)
  def troco(self, valDisponivel):
    return valDisponivel - self.valor
  def vender(self, valDisponivel):
    self.check_valor(valDisponivel)
    if valDisponivel >= self.valor:
      print(f"Você pode pagar o {self.tipo}")
    if valDisponivel == self.valor:
      print("É completo")
    else:
      print(f"troco R$ {self.troco(valDisponivel)}")
      exit("Thanks for your transaction")
```

```python
simples = Cafe('Café simples', 2)
cafeLeite = Cafe('Café com leite', 5)
expresso = Cafe('Expresso', 6)
try:
  valorPago = float(input('Qual seu valor disponível?: '))
except:
  exit('Digite um valor válido:')
for cafe in [expresso, cafeLeite, simples]:
  cafe.vender(valorPago)
```
